```python
#Read data
import pandas as pd

df = pd.read_excel (r'C:\Users\DILLEP J\Desktop\EQ06.xlsx',sheet_name='Data 1', skiprows=3) #this path is your your computer downloads from browser

df.head()
```

```python
#Checking with missing coloumns 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#I rename the columns to make it easier for me call the column names for future operations. And droping missing values data
df.columns.values[0] = 'Mid_quarter_month'
df.columns.values[2] = 'Greater capital city and rest of state'
df.columns.values[3] = 'Industry_division_of_main_job'
df.columns.values[4] = 'Employed_full_time'
df.columns.values[5] = 'Employed_part_time'
df.columns.values[6] = 'Number of hours actually worked in all jobs (full-time)'
df.columns.values[7] = 'Number of hours actually worked in all jobs (part-time)'
dfc = df.dropna(axis=1)
dfc
```

```python
#Converting categorical coloumns into numerical
df_numerized = dfc

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
#Number of people employed Full-time vs Part-time

current_df= dfc.iloc[-1]
labels = 'full-time', 'part-time'

sizes=[current_df['Employed_full_time'], current_df['Employed_part_time']]
custom_colours = ['#cf6a87', '#786fa6', '#f7d794']
#explode=(0.1, 0, 0)

plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, textprops={'fontsize':15}, autopct='%1.1f%%', shadow=False, 
        startangle=90, colors=custom_colours, pctdistance=0.8)

#draw circle
centre_circle = plt.Circle((0, 0), radius=0.5, fc='white')
plt.gca().add_artist(centre_circle)
plt.title('Percentage of Full-time, Part-time', size=20)
plt.show()
```

```python
#Numerized correlation matrix
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlarion matrix vs Numeric features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.show()
```

```python
#Employed full-time by Industry

# Prepare Data
dfN = dfc[['Employed_full_time', 'Industry_division_of_main_job']].groupby('Industry_division_of_main_job').apply(lambda x: x.mean())
dfN.sort_values('Employed_full_time', inplace=True)
#dfN.reset_index(inplace=True)
dfN.reset_index(drop=True, inplace=True)

# Draw plot
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(16,10), facecolor='white', dpi= 80)
ax.vlines(x=dfN.index, ymin=0, ymax=dfN.Employed_full_time, color='firebrick', alpha=0.7, linewidth=20)

# Annotate Text
for i, cty in enumerate(dfN.Employed_full_time):
    ax.text(i, cty+0.5, round(cty, 1), horizontalalignment='center')


# Title, Label, Ticks and Ylim
ax.set_title('Bar chart for full time employed by Industry ', fontdict={'size':22})
ax.set(ylabel='Industry', ylim=(0, 30))
dfN.astype(str).str.get(1)
plt.xticks(dfN.index, dfN.Industry_division_of_main_job.str.upper(), rotation=60, horizontalalignment='right', fontsize=12)

# Add patches to color the X axis labels
p1 = patches.Rectangle((.57, -0.005), width=.33, height=.13, alpha=.1, facecolor='green', transform=fig.transFigure)
p2 = patches.Rectangle((.124, -0.005), width=.446, height=.13, alpha=.1, facecolor='red', transform=fig.transFigure)
fig.add_artist(p1)
fig.add_artist(p2)
plt.show()
```
